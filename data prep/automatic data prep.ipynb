```python
def check_keys (data, ids):
    
    if (data[ids].isnull().sum().sum() == 0):
        print("Não há missings nos ids.")
    else:
        print("Checar possibilidade de missings nos ids.")
        
    check = data.groupby(ids).size().reset_index(name = 'count')
    check.sort_values('count', ascending=False)
    dup = check[check['count'] > 1]
    if (dup.shape[0] == 0):
        print('Sucesso! Base sem chaves duplicadas')
    else:
        print("ATENÇÃO: existem %d chaves duplicadas" % dup.shape[0] )
        print("\n")
        print(dup)
        
def var_type (x):
    x = str(x)
    if x.find('int') >= 0:
        return 'int'
    elif x.find('float') >=0 :
        return 'float'
    elif x.find('object') >= 0:
        return 'object'    
    else:
        return 'outros'
```

```python
def all_info_data (data, ids):
    
    #basic information
    print ('Data with', data.shape[0], 'samples and ', data.shape[1], 'features')
    #cheking duplicate keys
    print("\n")
    print("------- Informações sobre a chave --------\n")
    check_keys(data, ids)
    #about columns
    
    len_array = []
    for var in data.columns:
        current_len  = len(data[var].unique())
        len_array.append(current_len)

    missing = pd.DataFrame(np.array(data.isnull().sum()), columns = ['# missing'])
    
    p_missing = missing / data.shape[0]
    types = pd.DataFrame( np.array(data.dtypes), columns=['type'])

    qtd = pd.concat([pd.DataFrame(data.columns, columns=['var']), 
                     types.type, missing['# missing'], 
                     pd.DataFrame(p_missing),  
                     pd.DataFrame(len_array)], axis = 1)
    qtd.columns = ['var', 'type', '# missing','% missing','# unique values'] 
    qtd['type_resume'] = qtd['type'].apply(var_type)
    
    print("\n")
    print("-----Informações sobre as FEATURES ---------\n")
    #columns zero variance
    cols_zero = qtd[qtd['# unique values'] == 1].loc[:, 'var'].values
    if len(cols_zero) == 0:
        print('Não há colunas com a mesma informação.')
    else:
        print('ATENÇÃO: colunas contendo a mesma informação, devem ser excluídas posteriormente.')
        print(cols_zero)
        print("\n")
    
    n_sup_20_miss = len(list(filter(lambda x: x >= 0.2 , qtd['% missing'])))
    print ("Possui %d variáveis (%.1f%%) contendo mais de 20%% de dados faltantes" % (n_sup_20_miss, n_sup_20_miss*100/data.shape[1]))
        
    print("\n------Informações sobre as linhas ---------\n")
    
    #falar x% info completa
    
    data['vec_miss'] = data.isnull().sum(axis = 1)
    n_complete = len(list(filter(lambda x: x == 0, data['vec_miss'])))
    n_sup_80_mmiss = len (list(filter(lambda x: x/(data.shape[1]-1-len(ids)) > 0.8, data['vec_miss'])))
    
    print ("Possui %d das linhas (%.1f%%) contendo 100%% das variáveis\n" % (n_complete, n_complete*100/data.shape[0]))
    
    if n_sup_80_mmiss > 0:
        print ("ATENÇÃO: Possui %d das linhas (%.1f%%) contendo mais de 80%% de variáveis com dados faltantes\n" % (n_sup_80_mmiss, n_sup_80_mmiss*100/data.shape[0]))

    n_var = len(data.columns) - len(ids) - 1
    qtd_empty_rows = len(list(filter(lambda x: x == n_var, data['vec_miss'])))
    print('Dados com linhas vazias abaixo:')
    print(data[data.vec_miss == n_var ])
    
    # x% vazio tirando o ID :) deve ser excluida
    # possui x com 70% ou mais missing - segue indices abaixo.
    
    
    # dropar tudo que tiver direito
    
        
    
```

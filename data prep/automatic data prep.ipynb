```python
def check_keys (data, ids):
    check = data.groupby(ids).size().reset_index(name = 'count')
    check.sort_values('count', ascending=False)
    dup = check[check['count'] > 1]
    if (dup.shape[0] == 0):
        print('Sucesso! Base sem chaves duplicadas')
    else:
        print("ATENÇÃO: existem %d chaves duplicadas" % dup.shape[0] )
        print("\n")
        print(dup)
        
def var_type (x):
    x = str(x)
    if x.find('int') >= 0:
        return 'int'
    elif x.find('float') >=0 :
        return 'float'
    elif x.find('object') >= 0:
        return 'object'    
    else:
        return 'outros'
```
